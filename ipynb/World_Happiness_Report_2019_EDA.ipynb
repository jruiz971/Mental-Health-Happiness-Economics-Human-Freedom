```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Â¿QuÃ© paÃ­ses son los mÃ¡s felices?

```python
df_selected = df.iloc[np.r_[0:20, -20:0]]
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x=df_selected['Country or region'], y=df_selected['Score'],palette=sns.cubehelix_palette(len(df_selected)))
# Place the region names at a 90-degree angle.
plt.xticks(rotation= 90)
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('the 20 happiest countries and 20 unhappiest countries in the world')
plt.show()
```

## Paises con mayor GDP per cÃ¡pita

```python
#(df['Country or region'].head(10)) and (df['Country or region'].tail(10))
plt.figure(figsize=(12,10))
sns.barplot(x=df_selected['Country or region'], y=df_selected['GDP per capita'],palette=sns.cubehelix_palette(len(df_selected)))
# Place the region names at a 90-degree angle.
plt.xticks(rotation= 90)
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')
plt.title('GDP Per Capita of 20 happier countries vs Social Support of 10 unhappier countries')
plt.show()
```

```python
f,ax = plt.subplots(figsize = (9,5))
#sns.set_color_codes("pastel")
sns.barplot(x=countries_selected['Social support'],y=countries_selected['Country or region'],color='b',label="Social support")

#sns.barplot(x=countries_selected['Generosity'],y=countries_selected['Country or region'],color='c',label="Generosity")

sns.barplot(x=countries_selected['GDP per capita'],y=countries_selected['Country or region'],color='r',label="Economy")
sns.barplot(x=countries_selected['Healthy life expectancy'],y=countries_selected['Country or region'],color='g',label="Healty")
sns.barplot(x=countries_selected['Freedom to make life choices'],y=countries_selected['Country or region'],color='y',label="Freedom to make life choices")
sns.barplot(x=countries_selected['Perceptions of corruption'],y=countries_selected['Country or region'],color='k',label="Perceptions of corruption")
ax.legend(loc="lower right",frameon = True)
ax.set(xlabel='Percentage of Region', ylabel='Region',title = "Factors affecting happiness score")
plt.show()
```
